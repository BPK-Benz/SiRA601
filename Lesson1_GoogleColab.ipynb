```python
# Area of a Triangle (Using Heron‚Äôs Formula)
import math

a = 5
b = 6
c = 7
s = (a + b + c) / 2  # semi-perimeter

area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print(f"Area of the triangle: {area:.2f}")
```

```python
#Solve Quadratic Equation
# Equation: ax¬≤ + bx + c = 0 for a=1, b=-3, c=2
# (-b +/- b**2 - 4*a*c)/2a

import math

a, b, c = 1, -3, 2
discriminant = b**2 - 4*a*c

if discriminant >= 0:
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    print(f"Roots are real: {root1:.2f} and {root2:.2f}")
else:
    print("Complex roots", discriminant)
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        print("Result:", result)
    except ZeroDivisionError:
        print("Cannot divide by zero.\n")
    except TypeError:
        print("Please enter numbers only.\n")

for i, j in enumerate([0, "a", 4]):
    print(f"Case_{i + 1} Divisor: {j}")
    safe_divide(10, j)

```

```python
def test_common_errors():
    tests = [
        ("'abc' / 0", lambda: "abc" / 0),
        ("'5' / 0", lambda: "5" / 0),
        ("8 / 0", lambda: 8 / 0),
        ("[10, 20, 30][7]", lambda: [10, 20, 30][7]),
        ("dict['gender']", lambda: {"name": "Python"}["gender"]),
        ("undefined_var", lambda: undefined_var),
        ("'hello'.push('world')", lambda: "hello".push("world")),
        ("open('no_file.txt')", lambda: open("no_file.txt").read())
    ]
    for desc, func in tests:
        try:
            func()
        except Exception as e:
            print(f"{desc} ‚Üí {type(e).__name__}")


test_common_errors()

```
